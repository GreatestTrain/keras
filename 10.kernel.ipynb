```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img = io.imread('./input/yo.png')
img_gray = color.rgb2gray(img)
```

```python
subplot_kw={'xticks': [],
            'yticks': []}
```

```python
fig, ax  = plt.subplots(1,2,
                        subplot_kw=subplot_kw)
ax[0].imshow(img)
ax[0].set_title('imagen original')
ax[1].imshow(img_gray, cmap='binary_r')
ax[1].set_title('imagen a escala de grises')
plt.show()
```

```python
kernel = np.array([[-1,0,1], [-1,0,1], [-1,0,1]])
kernel_2 = kernel.T
kernel, kernel_2
```

```python
kernel_3 = np.array([[0.0625,0.125,0.0625],
                     [0.125,0.25,0.125],
                     [0.0625,0.125,0.0625]]) / 2 
```

```python
img_new = nd.convolve(img_gray, kernel)
img_new_2 = nd.convolve(img_gray, kernel_3)
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,6), subplot_kw=subplot_kw)
axes[0].imshow(img_gray, cmap='gray')
axes[1].imshow(img_new, cmap='gray')
axes[2].imshow(img_new_2, cmap='gray')
plt.show()
```
